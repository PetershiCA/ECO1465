```python
cat = survey[["ideologyF", "partyAllF", "trumpF", "genderF", "ideoLeaningF", "tookFollowup"]]
```

```python
dum = pd.get_dummies(cat, dtype=int)
dum
```

```python
nocat = survey[["ageWM", "thermo_DiffWM", "empathyDifficult_DiffWM", "matchTreatment"]]
```

```python
table = pd.concat([nocat, dum], axis=1)
table
```

```python
cols = table.columns.tolist()
cols.remove("matchTreatment")
```

```python
group_means = table.groupby('matchTreatment').mean().T
group_var = table.groupby('matchTreatment').var().T
group_counts = table.groupby('matchTreatment').count().T

m1 = group_means["Counter"].to_numpy()
m2 = group_means["Pro"].to_numpy()
v1 = group_var["Counter"].to_numpy()
v2 = group_var["Pro"].to_numpy()
n1 = group_counts["Counter"].to_numpy()
n2 = group_counts["Pro"].to_numpy()

mean_diff = m1 - m2
```

```python
n2
```

```python
t_stat = (m1 - m2) / np.sqrt(v1/n1 + v2/n2)

# Welch-Satterthwaite degrees of freedom
df_welch = (v1/n1 + v2/n2)**2 / ((v1/n1)**2 / (n1 - 1) + (v2/n2)**2 / (n2 - 1))

# Two-sided p-values
p_values = 2 * (1 - t.cdf(np.abs(t_stat), df_welch))


p_values
```

```python
balance_table = pd.concat([group_means[["Counter", "Pro"]], pd.Series(mean_diff, name = "Difference", index = group_means.index), pd.Series(p_values, name = "p-value", index = group_means.index)], axis=1)
balance_table
```
