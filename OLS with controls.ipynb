```python
model = smf.ols("polAffectiveIndex ~ C(matchTreatment)*thermo_DiffWM + C(matchTreatment)*empathyDifficult_DiffWM + C(ideologyF) + C(partyAllF) + C(trumpF) + C(genderF) + C(ageHave) + C(ideoLeaningF) + C(matchTreatment) + C(thermo_DiffHave) + C(empathyDifficult_DiffHave) + ageWM + ageSqWM + thermo_DiffWM + empathyDifficult_DiffWM", data=survey).fit()
```
