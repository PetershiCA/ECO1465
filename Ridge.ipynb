```python
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None)
```

```python
survey = pd.read_csv(r"C:\Users\Peter\OneDrive\Desktop\Code and Projects\ECO1465\clean survey.csv")
```

```python
survey = survey.dropna(subset=["polAffectiveIndex"])
survey
```

```python
X = survey[["ideologyF", "partyAllF", "trumpF", "genderF", "ageWM", "ageSqWM", "ageHave", "ideoLeaningF", "matchTreatment", 
            "thermo_DiffWM", "thermo_DiffHave", "empathyDifficult_DiffHave",
            "empathyDifficult_DiffWM"]]
Y = survey["polAffectiveIndex"]
```

```python
numeric_features = ["ageWM", "ageSqWM", "thermo_DiffWM", "empathyDifficult_DiffWM"]

categorical_features = ["ideologyF", "partyAllF", "trumpF", "genderF", "ideoLeaningF",
                        "ageHave", "thermo_DiffHave", "matchTreatment", "empathyDifficult_DiffHave",]  # one-hot encoded
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(drop='first'), categorical_features)
])
```

```python
ridge_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('ridge', Ridge())
])
```

```python
param_grid = {
    "ridge__alpha": np.linspace(0.01, 500, 50)  
}

cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# Grid search
grid_search = GridSearchCV(
    ridge_pipeline,
    param_grid,
    cv=cv,
    scoring='neg_mean_squared_error',
    return_train_score=True)

grid_search.fit(X, Y)
```

```python
# Best parameters
print("Best alpha:", grid_search.best_params_['ridge__alpha'])
print("Best CV RMSE:", np.sqrt(-grid_search.best_score_))
```

```python
# Extract results
results = grid_search.cv_results_
alphas = results['param_ridge__alpha'].data.astype(float)
mean_rmse = np.sqrt(-results['mean_test_score'])
std_rmse = results['std_test_score']
```

```python
results.keys()
```

```python
# Plot performance
plt.figure(figsize=(8, 5))
plt.plot(alphas, mean_rmse, marker='o', label="Mean CV RMSE")
plt.fill_between(alphas,
                 mean_rmse - std_rmse,
                 mean_rmse + std_rmse,
                 color='lightblue', alpha=0.5, label="1 std dev")

plt.xlabel("Alpha")
plt.ylabel("RMSE")
plt.title("Ridge Regression Cross-Validation Performance")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Extract coefficients for each alpha
coefs = []

alphas = np.linspace(0, 500, 50)

for alpha in alphas:
    ridge_pipeline.set_params(ridge__alpha=alpha)
    ridge_pipeline.fit(X, Y)
    # Get coefficients after preprocessing
    coefs.append(ridge_pipeline.named_steps['ridge'].coef_)

coefs = np.array(coefs)

feature_names = grid_search.best_estimator_.named_steps['preprocessor'].get_feature_names_out()

plt.figure(figsize=(8, 5))
for i, name in enumerate(feature_names):
    plt.plot(alphas, coefs[:, i], label=name)

plt.xlabel('Alpha')
plt.ylabel('Coefficient Value')
plt.title('Ridge Regression Coefficients vs Alpha')
plt.legend(loc='best', fontsize='xx-small', ncol=2)
plt.tight_layout()
plt.show()
```
